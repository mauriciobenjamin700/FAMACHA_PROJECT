````python
from ultralytics import YOLO
from pathlib import Path
import torch
````

````python
# Criando um modelo yolov8n do zero
model = YOLO(model= 'yolov8n.yaml')
````

````python
YAML = Path("data/data.yaml").resolve()
YAML.exists()
````

````python
torch.cuda.is_available()
````

````python
# Treinando o nosso modelo com o dataset previamente preparado usando a plataforma RoboFlow, se tiver GPU troque devide='cpu' para device=0,,patience=10
#results = model.train(data=YAML, epochs=50,batch=-1,imgsz=640,device="cuda",project="FAMACHA",pretrained=False,verbose=True,single_cls=True)
model.train(data=YAML, epochs=10,batch=-1,imgsz=640,device="cuda",project="FAMACHA",pretrained=False)
````

````python

# Evaluate the model's performance on the validation set
results = model.val()
````

````python
#se não gostarmos dos nossos resultados podemos re treinar
# Load a pretrained YOLO model (recommended for training)
#model = YOLO('yolov8n.pt')
````
